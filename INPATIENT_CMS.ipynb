```python
import pandas as pd
```

```python
data = pd.read_csv('https://data.cms.gov/sites/default/files/2023-04/67157de9-d962-4af0-bf0e-3578b3afec58/inpatient.csv', sep='|')
```

```python
data
```

```python
for i in data.columns:
  print(i)
```

```python
icd_columns=[i for i in data.columns if 'ICD' in i]
print(icd_columns)
```

```python
drg_columns=[i for i in data.columns if 'DRG' in i]
print(drg_columns)
```

```python
hcps_columns=[i for i in data.columns if 'HCPCS' in i]
print(hcps_columns)
```

```python
#Analyze the Frequency of Each Unique Value
#Calculate the frequency of unique values in each codex column
```

```python
ICD_DGNS_CD1_frequency=data['ICD_DGNS_CD1'].value_counts()
print(ICD_DGNS_CD1_frequency)
CLM_DRG_CD_frequency=data['CLM_DRG_CD'].value_counts()
print(CLM_DRG_CD_frequency)
HCPCS_CD_frequency=data['HCPCS_CD'].value_counts()
print(HCPCS_CD_frequency)
ICD_DGNS_CD3_frequency=data['ICD_DGNS_CD3'].value_counts()
print(ICD_DGNS_CD3_frequency)
```

```python
missing_icd_dgns_cd1=data['ICD_DGNS_CD1'].isnull().sum()
missing_clm_drg_cd=data['CLM_DRG_CD'].isnull().sum()
missing_hcpcs_cd=data['HCPCS_CD'].isnull().sum()
missing_icd_dgns_cd3=data['ICD_DGNS_CD3'].isnull().sum()
print(missing_icd_dgns_cd1)
print(missing_clm_drg_cd)
print(missing_hcpcs_cd)
print(missing_icd_dgns_cd3)
```

```python
data['CLM_DRG_CD'].fillna('NA',inplace=True)
data['ICD_DGNS_CD3'].fillna('NA',inplace=True)
```
